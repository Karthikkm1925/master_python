<strong style="font-size : 30px; color:green">Indexing</strong>

```python
# Create a 1D array
a = np.array([10,20,30,40,50])   
print("first element :",a[0])    # Access the first element

print("fourth element :",a[3])    # Access the fourth element

print("last element :",a[-1])   # Access the last element

print("second last element :",a[-2])   # Access the second last element
```

```python
# Create a 3x3 array
a = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])      
# Access the first row        
print("first row :",a[0])

# Access the second row    
print("second row :",a[1])

 # Access the third row    
print("third row :",a[2]) 

# Access the element at row 2, column 3
print("element at row 2, column 3 :",a[1][2]) 

# Access the element at row 3, column 2
print("element at row 3, column 2 :",a[2][1]) 

# Access the element at row 2, column 3 (recommended syntax)
print("element at row 2, column 3",a[1,2])  
```

<strong style="font-size : 30px; color:green">Slicing</strong>

```python
# 1D Array
# Create a 1D array
a = np.array([10,20,30,40,50,60,70]) 

# Access all elements
print("all elements :",a[:])   

# Access elements from index 2 to the end   
print("elements from index 2 to the end :",a[2:]) 

# Access elements from the beginning to index 4    
print("elements from the beginning to index 4 :",a[:5]) 

 # Access elements from index 2 to index 5  
print("from index 2 to index 5 :",a[2:6])   

# Access every second element
print("every second element :",a[::2]) 

# Reverse the array   
print("Reverse order :",a[::-1])   
```

```python
# 2D Array
# Create a 3x3 array
a = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])                
 # Access the second column
print("second column :",a[:,1])  

 # Access the second row
print("second row :",a[1,:]) 

# Access rows 1-2 and columns 2-3    
print("rows 1-2 and columns 2-3 :",a[0:2,1:3])  
```

<strong style="font-size : 30px; color:green">Reshape</strong>

```python
a = np.arange(12)        # Create an array with values from 0 to 11

print(a.reshape(3,4))    # Reshape into a 3x4 array
print(a.reshape(2,6))    # Reshape into a 2x6 array
print(a.reshape(4,3))    # Reshape into a 4x3 array
```

<strong style="font-size : 30px; color:green">Flatten</strong>

```python
 # Create a 2x3 array
a = np.array([
    [1,2,3],
    [4,5,6]
])                    
 # Convert the array into a new 1D array (creates a copy)
print(a.flatten())  
 # Convert the array into a 1D array (returns a view when possible)  
print(a.ravel())      
```

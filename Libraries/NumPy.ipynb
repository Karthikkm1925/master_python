<strong style="font-size : 25px; color:green">Indexing</strong>

```python
# Create a 1D array
a = np.array([10,20,30,40,50])   
print("first element :",a[0])    # Access the first element

print("fourth element :",a[3])    # Access the fourth element

print("last element :",a[-1])   # Access the last element

print("second last element :",a[-2])   # Access the second last element
```

```python
# Create a 3x3 array
a = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])      
# Access the first row        
print("first row :",a[0])

# Access the second row    
print("second row :",a[1])

 # Access the third row    
print("third row :",a[2]) 

# Access the element at row 2, column 3
print("element at row 2, column 3 :",a[1][2]) 

# Access the element at row 3, column 2
print("element at row 3, column 2 :",a[2][1]) 

# Access the element at row 2, column 3 (recommended syntax)
print("element at row 2, column 3",a[1,2])  
```

<strong style="font-size : 25px; color:green">Slicing</strong>

```python
# 1D Array
# Create a 1D array
a = np.array([10,20,30,40,50,60,70]) 

# Access all elements
print("all elements :",a[:])   

# Access elements from index 2 to the end   
print("elements from index 2 to the end :",a[2:]) 

# Access elements from the beginning to index 4    
print("elements from the beginning to index 4 :",a[:5]) 

 # Access elements from index 2 to index 5  
print("from index 2 to index 5 :",a[2:6])   

# Access every second element
print("every second element :",a[::2]) 

# Reverse the array   
print("Reverse order :",a[::-1])   
```

```python
# 2D Array
# Create a 3x3 array
a = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])                
 # Access the second column
print("second column :",a[:,1])  

 # Access the second row
print("second row :",a[1,:]) 

# Access rows 1-2 and columns 2-3    
print("rows 1-2 and columns 2-3 :",a[0:2,1:3])  
```

<strong style="font-size : 25px; color:green">Arithmetic</strong>

```python
a = np.array([10,20,30])    # Create the first array
b = np.array([1,2,3])       # Create the second array

print(a + b)    # Add corresponding elements
print(a - b)    # Subtract corresponding elements
print(a * b)    # Multiply corresponding elements
print(a / b)    # Divide corresponding elements
print(a // b)   # Perform floor division on corresponding elements
print(a % b)    # Find the remainder of corresponding elements
print(a ** 2)   # Square each element of array 'a'
```

<strong style="font-size : 25px; color:green">Universal Functions</strong>

```python
# Create a NumPy array
a = np.array([4,9,16,25])   

print(np.sqrt(a))     # Calculate the square root of each element
print(np.square(a))   # Calculate the square of each element
print(np.exp(a))      # Calculate e raised to the power of each element
print(np.log(a))      # Calculate the natural logarithm (base e) of each element
print(np.sin(a))      # Calculate the sine of each element (in radians)
print(np.cos(a))      # Calculate the cosine of each element (in radians)
print(np.tan(a))      # Calculate the tangent of each element (in radians)
```

<strong style="font-size : 25px; color:green">Statistics</strong>

```python
 # Create a NumPy array
a = np.array([10,20,30,40,50])  

print(np.sum(a))      # Calculate the sum of all elements
print(np.mean(a))     # Calculate the average (mean) of the elements
print(np.median(a))   # Calculate the median (middle value)
print(np.max(a))      # Find the maximum element
print(np.min(a))      # Find the minimum element
print(np.std(a))      # Calculate the standard deviation
print(np.var(a))      # Calculate the variance
print(np.prod(a))     # Calculate the product of all elements
```

<strong style="font-size : 25px; color:green">Axis Operations</strong>

```python
# Create a 2D NumPy array
a = np.array([
    [10,20,30],
    [40,50,60]
])                        

print(np.sum(a))          # Calculate the sum of all elements
print(np.sum(a,axis=0))   # Calculate the column-wise sum
print(np.sum(a,axis=1))   # Calculate the row-wise sum
print(np.mean(a,axis=0))  # Calculate the column-wise average
print(np.mean(a,axis=1))  # Calculate the row-wise average
```

<strong style="font-size : 25px; color:green">Copy vs View</strong>

```python
# Create an original array
a = np.array([1,2,3])        
# Create a separate copy of the array
b = a.copy()                 
# Create a view sharing the same memory
c = a.view()                 
# Modify the first element of original array
a[0] = 100                   
# Display original array
print(a)                    
# Display copied array (unchanged)
print(b)    
# Display view array (changed)                 
print(c)                     
```

<strong style="font-size : 25px; color:green">Broadcasting</strong>

```python
# Create a 2D array
a = np.array([
    [1,2,3],
    [4,5,6]
])                              
# Add a scalar value to every element
print(a + 10)                   
# Create a 1D array
b = np.array([100,200,300])    
# Add array values using broadcasting
print(a + b)                   
```

<strong style="font-size : 25px; color:green">Matrix Multiplication</strong>

```python
# Create first matrix
a = np.array([
    [1,2],
    [3,4]
])                              
# Create second matrix
b = np.array([
    [5,6],
    [7,8]
])                              
# Perform matrix multiplication using @ operator
print(a @ b)                    
# Perform matrix multiplication using dot function
print(np.dot(a,b))              
```

<strong style="font-size : 25px; color:green">Linear Algebra</strong>

```python
# Create a square matrix
a= np.array([
    [1,2],
    [3,4]
])                         
# Calculate the determinant of the matrix
print("Determinant :",np.linalg.det(a),"\n")     
# Calculate the inverse of the matrix
print("Leverage :",np.linalg.inv(a),"\n")     
# Calculate eigenvalues and eigenvectors of the matrix
print("Eigenvalues and Eigenvector :",np.linalg.eig(a))     
```

<strong style="font-size : 25px; color:green">Helpful Functions</strong>

```python
a  = np.arange(12)           # Create an array with values from 0 to 11

print(a.reshape(3,4))       # Reshape array into 3 rows and 4 columns

print(a.reshape(-1,2))      # Automatically calculate rows with 2 columns

print(a.reshape(2,-1))      # Automatically calculate columns with 2 rows

print(a.argmax())           # Find the index position of maximum value

print(a.argmin())           # Find the index position of minimum value

print(np.clip(a,3,8))       # Limit values between minimum 3 and maximum 8

print(np.round(3.14159,2))  # Round value to 2 decimal places

print(np.ceil(3.2))         # Round value upward to the next integer

print(np.floor(3.9))        # Round value downward to the previous integer
```

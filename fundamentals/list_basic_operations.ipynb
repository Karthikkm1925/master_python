<strong style="font-size : 30px; color:green">List In Python</strong>

<strong style="font-size : 20px; color:green">What is a List in Python?</strong>

```python
#List is an Ordered, mutable collection that stores multiple values
fruits = ["Apple","Orange","Mango"]
print(fruits)
```

<strong style="font-size : 25px; color:green">Creating List</strong>

```python
#Creating Lists: Create a list using square brackets []
nums = [1,2,3,4,5]
names = ["Alan","Jahk","John"]
empty = []
print(nums)
print(names)
print(empty)
```

<strong style="font-size : 25px; color:green">Accessing Elements (Indexing)</strong>

```python
#Indexing means Access list elements using their position index
fruits = ["apple", "banana", "mango"]
print(fruits[0])   # apple
print(fruits[1])   # banana
```

<strong style="font-size : 25px; color:green">Negative Indexing</strong>

```python
print(fruits[-1])  # mango
print(fruits[-2])  # banana
```

<strong style="font-size : 25px; color:green">Slicing Lists</strong>

```python
#Slicing means Extract a portion of a list using start:end:step,  syntax -> list[start:end:step]
nums = [10,20,30,40,50]
print(nums[1:4])   # [20,30,40]
print(nums[:3])    # [10,20,30]
print(nums[2:])    # [30,40,50]
print(nums[::-1])  # reverse
```

<strong style="font-size : 25px; color:green">Changing Elements</strong>

```python
#Changing Elements means Modify list items by assigning new values
fruits = ["apple", "banana", "mango"]
fruits[1] = "orange"
print(fruits)
```

<strong style="font-size : 25px; color:green">Adding Elements using append(), insert(), extend()</strong>

```python
#Adding Elements - Add items using append(), insert(), or extend()
#append() - Adds one item at the end.
list1 = [1,2,3]
list1.append(4)
print(list1)

#insert() - Adds item at specific index
list2 = [1,2,4]
list2.insert(2,5)
print(list2)

#extend - Adds multiple items.
list3 = [1,2]
list3.extend([3,4,5])
print(list3)
```

<strong style="font-size : 25px; color:green">Removing Elements using remove(), pop(), clear()</strong>

```python
#Removing Elements: Remove items using remove(), pop(), del, or clear()

#remove() - Removes first matching value
l1 = [1,2,3,2]
l1.remove(2)
print(l1)

#pop(index) - Removes element by index
l2 = [10,20,30]
l2.pop(1)
print(l2)

#pop() - Pop Last Element
l2.pop()
print(l2)

#del(index) - delete specic index value, or entire list, or range of value with slicing  
l3 = [10,20,30]
del l3[1]
print(l3)

#clear() - removes all the elements in list
l4 = [1,2,3]
l4.clear()
print(l4)
```

<strong style="font-size : 25px; color:green">List Length</strong>

```python
#List Length - Use len() to find the number of elements
l1 = [1,2,3,4,5]
print(len(l1))
```

<strong style="font-size : 25px; color:green">Membership Operators</strong>

```python
#Membership Operators - Check if an item exists using in or not in
fruits = ["apple","banana","mango"]
print("banana" in fruits)
print("grapes" not in fruits)
```

<strong style="font-size : 25px; color:green">List Concatenation</strong>

```python
#List Concatenation: Combine multiple lists using the (+) operator
a = [1,2]
b = [3,4]
c = a + b
print(c)
```

<strong style="font-size : 25px; color:green">List Repetition</strong>

```python
#List Repetition: Repeat list elements using the * operator
nums = [1,2]
print(nums * 3)
```

<strong style="font-size : 25px; color:green">List Useful Built-in Methods index(), count(), sort(), reverse(), copy()</strong>

```python
# index(): Returns the position of the first matching element.
l1 = [10,20,30]
print(l1.index(20))

# count(): Counts how many times a value appears in a list.
l2 = [1,2,2,2,3]
print(l2.count(2))

# sort(): Sorts list elements in ascending or descending order.
#Acending Order
l3 = [5,2,8,1]
l3.sort()
print(l3)

#Decending Order
l3.sort(reverse=True)
print(l3)
# reverse(): Reverses the order of elements in the list.
nums = [1,2,3]
nums.reverse()
print(nums)
# copy(): Creates a separate copy of an existing list.
a = [1,2,3]
b = a.copy()
print(b)
```

<strong style="font-size : 25px; color:green">Looping Through Lists</strong>

```python
#Looping Through Lists - Iterate over list elements using loops

#Using for loop
print("Using for loop:")
fruits = ["apple","banana","mango"]
for fruit in fruits:
    print(fruit)

print("\n")

#Using range()
print("Using Range:")
for i in range(len(fruits)):
    print(fruits[i])
```

<strong style="font-size : 25px; color:green">List Comprehension</strong>

```python
#List Comprehension: Create lists efficiently in a single line

#Traditional way
square1 = []
for i in range(1,6):
    square1.append(i*i)
print(square1)

#List Comprehension
square2 = [i*i for i in range(1,6)]
print(square2)
```

<strong style="font-size : 25px; color:green">Nested Lists</strong>

```python
#Nested Lists: Store lists inside another list
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
print(matrix[1][2])
```

<strong style="font-size : 25px; color:green">max(), min(), sum(), sorted(list_name)</strong>

```python
nums = [10,20,30,40]
print(min(nums))   # 10
print(max(nums))   # 40
print(sum(nums))   # 100
print(sorted(nums))
```

<strong style="font-size : 25px; color:green">Shallow Copy vs Assignment</strong>

```python
#Shallow Copy vs Assignment - Understand reference sharing and copying
a = [1,2,3]
b = a
b.append(4)
print(a)
```

<strong style="font-size : 25px; color:green">Creating Dictionary</strong>

```python
#Dictionary is a Data Structure in pyhton used to store data as a key-value pair
#empty dictionary
empty = {}
print(empty)
#Dictionary with key Value Pairs
mydictionary = {1:"One",2:"Two"}
print(mydictionary)
```

<strong style="font-size : 25px; color:green">Accessing Dictionary Values</strong>

```python
#If key does not exist
dict1 = {
    "Name":"Alan",
    "Age":25
}
print(dict1["Age"])
```

<strong style="font-size : 25px; color:green">Accessing Dictionary Values Using get()</strong>

```python
#safer if key is missing, gives output None
dict2 ={
    "Name":"Alan",
    "Age":25
}
print(dict2.get("Gnder"))

#Custom default:
print(dict2.get("Gender","Not Available"))
```

<strong style="font-size : 25px; color:green">Adding New Values</strong>

```python
dict3 = {
    "Name":"Alan",
    "Age":25
}
dict3["Gender"]="Male"
print(dict3)
```

<strong style="font-size : 25px; color:green">Updating Existing Values</strong>

```python
dict4 = {
    "Name":"Alan",
    "Age":25
}
print(dict4)

#Update value
dict4["Age"]=26
print(dict4)
```

<strong style="font-size : 25px; color:green">Update Dictionary using update()</strong>

```python
dict5 = {
    "Name":"Alan",
    "Age":26
}
dict5.update({
    "Gender":"Male",
    "Age":27
})
print(dict5)
```

<strong style="font-size : 25px; color:green">Removing Dictionary Elements</strong>

```python
#pop() : Removes specific key
print("Demo of using pop()")
dict6 = {
    "Name":"Alan",
    "Gender":"Male",
    "Age":25
}
dict6.pop("Gender")
print(dict6)
print(dict6.pop("Age"))
print(dict6)
print("\nDemo of using popitem()")
#popitem() : Removes last inserted item
dict7 = {
    "Name":"Alakrita",
    "Age":22,
    "Gender":"Female"
}
print(dict7)
print(dict7.popitem())

#del : used to delete particular item or to delete entire dictionary
print("\nDemo of using del for specific item")
dict8 ={
    "Name":"John",
    "Country":"USA",
    "Age":27,
}
del dict8["Age"]
print(dict8)
#print("Demo of using del to remove entire dictionary")
del dict8
#print(dict8) #gives error saying dict8 not defined
```

```python
#clear(): Remove all items only not dictionary
dict9 = {
    "Name":"Antony",
    "Age":28
}
print("Demo of clear()")
dict9.clear()
print(dict9)
```

<strong style="font-size : 25px; color:green">Dictionary Length using len()</strong>

```python
dict10 = {
    "Name":"Anamenon",
    "Gender":"Female"
}
print(dict10)
print(len(dict10))
```

<strong style="font-size : 25px; color:green">Dictionary Methods keys(), values(), items()</strong>

```python
dict11 = {
    "Name":"Alan",
    "Age":28
}
#keys() returns all keys
print(dict11.keys())
#values() returns all values
print(dict11.values())
#items() returns all items, key-value pair
print(dict11.items()) #without print() will get output for items()
```

<strong style="font-size : 25px; color:green">Check key in dictionary</strong>

```python
dict12 = {
    "Name":"Mark",
    "Age":28,
    "Gender":"Male"
}
#key exist?
print("Name" in dict12)
```

<strong style="font-size : 25px; color:green">Looping through key, value, key-value in dictionary</strong>

```python
dict13 = {
    "Name":"Antony",
    "Country":"UK"
}
print("Loop through keys")
for key in dict12 :
    print(key)
print("\nLoop through Values")
for value in dict12.values() :
    print(value)
print("\nLoop through items (key-value)")
for key,value in dict12.items():
    print(key,value)
```

<strong style="font-size : 25px; color:green">Nested Dictionary</strong>

```python
students = {

    "student1":{
        "name":"John",
        "age":25
    },

    "student2":{
        "name":"Razina",
        "age":22
    }
}

print(students)
print(students["student1"]["name"])
```

<strong style="font-size : 25px; color:green">Dictionary with List</strong>

```python
player = {
    "name":"Pardeep",
    "points":[2,3,4],
    "matches":10
}
print(player["points"][1])
```

<strong style="font-size : 25px; color:green">Copying Dictionary</strong>

```python
#Wrong way
a1 = {
    "name":"Antony"
}

b1 = a1

b1["name"]="Alex"

print(a1)

#Correct way
a2 = {
    "name":"John"
}

b2 = a2.copy()

b2["name"]="Jack"

print(a2)
```

<strong style="font-size : 25px; color:green">Dictionary Comprehension</strong>

```python
#Creating dictionary using loop
number1 = {}

for i in range(1,6):
    number1[i]=i*i

print(number1)

#Using comprehension
number2 = {
    i:i*i
    for i in range(1,6)
}

print(number2)
```

<strong style="font-size : 25px; color:green">Creating Set</strong>

```python
'''
1. Set stores unique values only
2. Set has no index
3. Set is mutable
4. Set elements must be immutable
5. Order is not guaranteed
6. Duplicate values are removed automatically
7. Used for fast searching and removing duplicates
'''
set1 = set()
print("This is an Empty set",set1)
print(type(set1))
#Set with values
set2 = {1,2,3,4}
print("Set with values",set2)
#Duplicate Values Automatically Removed from set
set3 = {1,1,1,2,2,3}
```

<strong style="font-size : 25px; color:green">Adding Elements to Set using add(), update()</strong>

```python
#add() : appends the new element at the end
set1 = {1,2,3,4}
set1.add(5)
print(set1)
#update() : appends multiple elements at the end
set1.update([6,7,8])
print("After Update set values\n",set1)
```

<strong style="font-size : 25px; color:green">Remove Elements from Set using remove(), discard(), pop(), clear()</strong>

```python
#remove() : remove specified element from set, throughs error if element not exist
set2 = {10,20,30,40,50}
print(set2)
set2.remove(20)
print("After removing an value, remining elements are\n",set2)

#discard() : Safe delete option, removes if specified element present
set2.discard(30)
print("After discarding a value\n",set2)

#pop() : removes a random element
set2.pop()
print("After poping a value\n",set2)

#clear() is used to remove all the elements from set
set2.clear()
print("Empty",set2)
```

<strong style="font-size : 25px; color:green">Checking length of the set using len()</strong>

```python
set3 = {1,2,3,4,5,6}
print("Length of the set is",len(set3))
```

<strong style="font-size : 25px; color:green">Check value exist in given set using 'in' operator</strong>

```python
set4 = {1,2,3,4}
print(4 in set4)
```

<strong style="font-size : 25px; color:green">Loop through set </strong>

```python
set5 = {1,2,3,4,5,7,8,9}
for i in set5:
    print(i, end=" ")
```

<strong style="font-size : 25px; color:green">Set Operators</strong>

```python
set1 = {1,2,3,4,5,7}
set2 = {7,9,8,6,1,2,3}
#Union " | ": combines 2 sets
print(set1 | set2) 
#Intersection " & " : Retains common values from 2 sets
print(set1 & set2)
#Difference " - " : set1 - set2 
print(set1 - set2)
#Reverse Difference set2 - set1
print(set2-set1)
#Symmetric Difference
print(set1^set2)
```

<strong style="font-size : 25px; color:green">Set Methods union(), intersection(), difference(), symmetric_difference()</strong>

```python
set1 = {1,2,3,4,5}
set2 = {3,4,5,6,7}
#union()
print(set1.union(set2))
#intersection()
print(set1.intersection(set2))
#difference()
print(set1.difference(set2))
#symmetric difference
print(set1.symmetric_difference(set2))
```

<strong style="font-size : 25px; color:green">Subset</strong>

```python
#Check if one set exists inside another.
A = {1,2}
B = {1,2,3,4}
print(A.issubset(B))
```

<strong style="font-size : 25px; color:green">Superset</strong>

```python
A = {1,2}
B = {1,2,3,4}
print(B.issuperset(A))
```

<strong style="font-size : 25px; color:green">Frozen Set</strong>

```python
#Cannot modify.
a = frozenset([1,2,3])
print(a)
```

<strong style="font-size : 25px; color:green">Converting List to Set and Set List, vice-versa </strong>

```python
#List to Set : useful to remove duplicate 
#Useful for removing duplicates.
numbers = [1,2,2,3,3,4]
unique = set(numbers)
print(unique)

#Set to List
a = {1,2,3}
b = list(a)
print(b)
```

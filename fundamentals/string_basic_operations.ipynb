<strong style="font-size:30px; color:green">Basic operations on Strings In Python</strong>

<strong style="font-size:25px; color:green">Creating Strings</strong>

```python
# String are group of charecters, represented between quotes
# \(charecter) -> treated as escape charecter
print("\"This is String in doublwe quotes\"")
print('\'String in single quote\'')
print('''\'\'\'This is 
multi line 
string
in single quotes
\'\'\'''')
print("""\"\"\"
This is multi line string
represented in double quotes
\"\"\"""")
```

<strong style="font-size:25px; color:green">Accessing Characters (Positive Indexing)</strong>

```python
#String Creation - Store text data
text = "Programming"
#Indexing - Access single character
#Python uses 0-based indexing.
print(text[0]) # p
print(text[1]) # r
print(text[2]) # o
```

<strong style="font-size:25px; color:green">Negative Indexing</strong>

```python
#Negative Indexing - Access from end
text = "Programming"
print(text[-1]) # g
print(text[-2]) # n
```

<strong style="font-size:25px; color:green">String Length using len()</strong>

```python
#Length - Count characters using lent()
text = "Larger"
print(len(text))
```

<strong style="font-size:25px; color:green">String Slicing</strong>

```python
#Slicing - Get part of string using syntax, 
# syntax -> string[start:end:step]
text = "python"
print(text[0:3]) # pyt
print(text[2:]) # thon
print(text[:4]) # pyth
print(text[-3:]) # hon
```

<strong style="font-size:25px; color:green">Reverse String</strong>

```python
#Reverse String - Reverse using slicing
text = "Python"
print(text[::-1])
```

<strong style="font-size:25px; color:green">String Concatenation (+)</strong>

```python
#Concatenation - Join strings or string with variable value
name = "Antony"
print("Hello "+name)
```

<strong style="font-size:25px; color:green">String Repetition (*)</strong>

```python
#Repetition - Repeat string
print("Hi "*4)
```

<strong style="font-size:25px; color:green">Membership Operators (in)</strong>

```python
#Membership - Check existence and return True or False
print("Py" in "Python")
```

<strong style="font-size:25px; color:green">String Traversal Using for loop</strong>

```python
#Traversal - Iterate through string
text = "python"
for ch in text:
    print(ch, end=" ")
```

<strong style="font-size:25px; color:green">String Traversal Using index</strong>

```python
text = "Python"
for i in range(len(text)):
    print(i,text[i])
```

<strong style="font-size:30px; color:green">String Methods</strong>

<strong style="font-size:25px; color:green">upper(), lower(), title()</strong>

```python
text = "PyThoN pRoGramminG"
print(text.upper()) # conver string to Upper
print(text.lower()) # convrt string to lower
print(text.capitalize()) # convert first charecter string to Capital 
print(text.title()) # convert every first letter of each word to capital
```

<strong style="font-size:25px; color:green">replace()</strong>

```python
text = "C++ is easy to learn"
print(text.replace("C++","Python")) # replace the particular string with given string
```

<strong style="font-size:25px; color:green">lstrip(), rstrip(), strip()</strong>

```python
text = "    Python is Programming Language      "

print(text.lstrip())  # Removes leading spaces from the string
print(text.rstrip())  # Removes trailing spaces from the string
print(text.strip())   # Removes both leading and trailing spaces from the string
```

<strong style="font-size:25px; color:green">split()</strong>

```python
sentence = "Python beginner friendly to learn"
word = "Hello"
print(sentence.split()) #splits string into List values
print(word.split())
```

<strong style="font-size:25px; color:green">join()</strong>

```python
words = ["Hello", "Guys", "Learn", "Coding"]
print(" ".join(words)) #concatinates list to string
```

<strong style="font-size:25px; color:green">find()</strong>

```python
text = "Programming"
#returns first(start) index of the substring to find
print(text.find("gram"))  
#if not present return -1
print(text.find("omg"))
```

<strong style="font-size:25px; color:green">count()</strong>

```python
# count the occurance of substring (single charecter is also substring)
text = "Python"
print(text.count("t")) 
```

<strong style="font-size:25px; color:green">startswith(), endswith()</strong>

```python
text = "Extraordinary"
 #check string satrts with given substring return True or False
print(text.startswith("Extra")) 
#check string ends with given substring return True or False
print(text.endswith("ry")) 
```

<strong style="font-size:25px; color:green">islower(), isupper(), isdigit(), isalpha(), isalnum()</strong>

```python
#check if string has only lowercase
print("text".islower())
#check if string has only uppercase
print("TEXT".isupper())
#check if string has only digit 
print("123".isdigit()) 
#check if string has only alphabets
print("Hello".isalpha())
#check if string has both digit and alphabets
print("Hello123".isalnum())
```

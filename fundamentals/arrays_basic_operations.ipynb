<strong style="font-size : 30px; color:green">Arrays in Python</strong>

<strong style="font-size : 25px; color:green">Creating Arrays Using Lists (Python's Dynamic Array)</strong>

```python
#An array is a collection of elements stored in a sequence.
#array_name = [value1, value2, value3]
numbers = [10, 20, 30, 40, 50]
print(numbers)
```

<strong style="font-size : 25px; color:green">Accessing Array Elements (Positive Indexing)</strong>

```python
numbers = [10, 20, 30, 40, 50]
print(numbers[0])
print(numbers[3])
```

<strong style="font-size : 25px; color:green">Negative Indexing</strong>

```python
numbers = [10,20,30,40,50]
print(numbers[-1])
print(numbers[-3])
```

<strong style="font-size : 25px; color:green">Updating Array Elements</strong>

```python
#Arrays are mutable.
numbers = [10,20,30,40]
numbers[2] = 100
print(numbers)
```

<strong style="font-size : 25px; color:green">Adding Elements using append()</strong>

```python
#append() adds element at the end
numbers = [10,20,30]
numbers.append(40)
print(numbers)
```

<strong style="font-size : 25px; color:green">insert()</strong>

```python
#Adds element at a specific position.
numbers = [10,20,30]
numbers.insert(1,100)
print(numbers)
```

<strong style="font-size : 25px; color:green">extend()</strong>

```python
#Adds multiple elements
numbers = [10,20,30]
numbers.extend([40,50,60])
print(numbers)
```

<strong style="font-size : 25px; color:green">remove()</strong>

```python
#Removing Elements
numbers = [10,20,30,40]
numbers.remove(30)
print(numbers)
```

<strong style="font-size : 25px; color:green">pop()</strong>

```python
#Removes and returns value.
numbers = [10,20,30]
x = numbers.pop()
print(x)
print(numbers)

#Removes using index and returns value.
numbers.pop(0)
```

<strong style="font-size : 25px; color:green">del</strong>

```python
#Deletes element permanently.
numbers = [10,20,30]
del numbers[1]
print(numbers)
```

<strong style="font-size : 25px; color:green">len()</strong>

```python
numbers = [10,20,30,40]
print(len(numbers))
```

<strong style="font-size : 25px; color:green">Looping Through Arrays</strong>

```python
#Using for loop
numbers = [10,20,30,40]
for x in numbers:
    print(x)

#Using index
numbers = [10,20,30]
for i in range(len(numbers)):
    print(numbers[i])
```

<strong style="font-size : 25px; color:green">Array Slicing</strong>

```python
#array[start:end]
numbers = [10,20,30,40,50]
print(numbers[1:4])

#Start omitted
print(numbers[:3])

#end omitted
print(numbers[2:])

#Step slicing
print(numbers[::2])
```

<strong style="font-size : 25px; color:green">Copying Arrays using copy()</strong>

```python
#Wrong way, because both point to same address
ar = [1,2,3,4]
ab = ar
print(ar)
print(ab)
#correct way of coping
a=[1,2,3]
b=a.copy()
b[0]=100
print(a)
#Slicing copy
b=a[:]
print(b)
```

<strong style="font-size : 25px; color:green">Searching in Array using 'in'</strong>

```python
numbers=[10,20,30]
print(20 in numbers)
```

<strong style="font-size : 25px; color:green">Using index()</strong>

```python
#Returns position.
numbers=[10,20,30]
print(numbers.index(20))
```

<strong style="font-size : 25px; color:green">Counting Elements using count()</strong>

```python
numbers=[10,20,20,30]
print(numbers.count(20))
```

<strong style="font-size : 25px; color:green">Sorting Arrays</strong>

```python
#Ascending
numbers=[40,10,30,20]
numbers.sort()
print(numbers)

#Descending
numbers.sort(reverse=True)
print(numbers)

#using reverse()
numbers=[10,20,30]
numbers.reverse()
print(numbers)
```

<strong style="font-size : 25px; color:green">Two-Dimensional Arrays (Matrix)</strong>

```python
#Array inside array.
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

print(matrix)
#Access
#matrix[row][column]
print(matrix[1][2])
```

<strong style="font-size : 25px; color:green">Traversing 2D Array</strong>

```python
matrix=[
[1,2,3],
[4,5,6]
]

for row in matrix:
    for value in row:
        print(value)
```

<strong style="font-size : 25px; color:green">Array Comprehension</strong>

```python
numbers=[i for i in range(5)]
print(numbers)
```

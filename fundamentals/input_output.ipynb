<strong style="font-size : 30px; color:green">Input and Output Function</strong>

<strong style="font-size:25px; color:green">Standard input using input()</strong>

```python
#Standard input using input() function
variabl_name = input("Enter some text as input :\n") #input() function returns a string
print("Text You Entered is :\n",variabl_name) #Basic output function print()
```

<strong style="font-size:25px; color:green">Input type conversion to number using int()</strong>

```python
#input() returns string need conversion, int() converts strint to number
#Number input
number = int(input("Enter a number :\n")) #Number input
print("The Number you Entered is :",number) #output
```

<strong style="font-size:25px; color:green">Multiple inputs in one line using split()</strong>

```python
num1,num2 = input("Enter two numbers : ").split()
print("The Two Numbers you Entered are :",num1,",",num2)
```

<strong style="font-size:25px; color:green">Multiple inputs in one line With type conversion using map()</strong>

```python
num1,num2,num3 = map(int, input("Enter three numbers : ").split())
print("Entered Numbers are :",num1,num2,num3)
```

<strong style="font-size:25px; color:green">Taking list input using list()</strong>

```python
numbers = list(map(int, input("Enter List of Numbers :").split()))
print("The List of Numbers you Entered are :\n",numbers)
```

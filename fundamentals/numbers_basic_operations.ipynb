<strong style="font-size : 30px; color:green">Basic Operations on Numbers in Python</strong>

<strong style="font-size : 25px; color:green">Integer (int)</strong>

```python
#Whole numbers without decimal points.
num_1 = 10
num_2 = -11
num_3 = 0
print("Integers","\nPositive Integer :",num_1,"\nNegative Integer :",num_2,"\nZero :",num_3)
```

<strong style="font-size : 25px; color:green">Float (float)</strong>

```python
#Numbers with decimal points.
float_value = 3.142
print("Decimal Number :",float_value)
```

<strong style="font-size : 25px; color:green">Complex Numbers (complex)</strong>

```python
#numbers with real and imaginary parts
complex_num = 2 + 2j
print("Complex Number :",complex_num)
```

<strong style="font-size : 25px; color:green">Check Number Type using type(variable_name)</strong>

```python
#Finds the data type of a value.
num_1 = 100
num_2 = 100.0
num_3 = 100 + 100j
print(type(num_1))
print(type(num_2))
print(type(num_3))
```

<strong style="font-size : 30px; color:green">Type Conversion</strong>

<strong style="font-size : 25px; color:green">Integer to Float</strong>

```python
#Changes a value from one data type to another
num = 10
print(float(num))
```

<strong style="font-size : 25px; color:green">Float to Integer</strong>

```python
num = 10.11
print(int(num))
```

<strong style="font-size : 25px; color:green">String to Integer</strong>

```python
num = "100"
print(int(num))
```

<strong style="font-size : 25px; color:green">Arithmetic Operators</strong>

```python
num_1 = 3
num_2 = 2
print("Addition :",num_1+num_2)
print("Subtraction :",num_1-num_2)
print("Multiplication :",num_1*num_2)
print("Division :",num_1/num_2)
print("Integer or Flore Division :",num_1//num_2)
print("Modulus or Reminder :",num_1%num_2)
print("Exponent or Power :",num_1**num_2)
```

<strong style="font-size : 25px; color:green">Absolute Value using abs(value)</strong>

```python
#Returns the positive value of a number
print("Absolute value of ",-24,"is :",abs(-24))
```

<strong style="font-size : 25px; color:green">Round Number using round(value)</strong>

```python
#Rounds a number to the nearest specified value
print(round(3.45678)) #4
print(round(3.456,2)) #3.46
print(round(3.4)) #3
```

<strong style="font-size : 25px; color:green">Maximum, Minimum, Sum</strong>

```python
#Finds the largest or smallest value and the sum
nums = [10,20,30]
print("Max value :",max(nums))
print("Min value :",min(nums))
print("Sum of values :",sum(nums))
```

<strong style="font-size : 25px; color:green">Mathematical Functions</strong>

```python
import math
print(math.sqrt(4)) #return square root
print(math.ceil(4.4)) #round off to higer value
print(math.floor(4.5)) #round off to lower value
```

<strong style="font-size : 25px; color:green">Random Number</strong>

```python
#Generates random Integer number in given range
import random
#for loop to generate multiple times
for i in range(1,10):
    print(random.randint(100000,999999))
```

<strong style="font-size : 25px; color:green">Binary, Octal, Hexadecimal</strong>

```python
print(bin(100)) #binary representation of Integer number
print(oct(100)) #Octal representation
print(hex(100)) #Hexadecimal representation
```

<strong style="font-size : 25px; color:green">Number Formatting</strong>

```python
#Number Formatting means displaying a number in a more readable or desired format without changing its actual value

#Comma Separator
salary = 1000000
print(f"{salary:,}")

#Decimal Places
pi = 3.14159265
print(f"{pi:.2f}") #.2f means show 2 digits after the decimal point

#Leading Zeros
num = 7
print(f"{num:03}") #add required number of zeroes in prefix
```

<strong style="font-size : 30px; color:green">Tuple In Python</strong>

<strong style="font-size : 20px; color:green">Creating an empty tuple</strong>

```python
# Tuple is an ordered and immutable collection of elements
# Creating an empty tuple
t = ()
print(t)

# Creating a tuple with multiple values
t = (10, 20, 30)
print(t)

# Tuple can store different data types
t = (10, "Python", 5.5, True)
print(t)

# Parentheses are optional while creating tuples
t = 10, 20, 30
print(t)

# Single element tuple requires a comma
t = (10,)
print(type(t))
```

<strong style="font-size : 25px; color:green">Accessing tuple elements (Positive indexing)</strong>

```python
t = (10, 20, 30)
print(t[0])
```

<strong style="font-size : 25px; color:green">Accessing tuple elements (Negative  indexing)</strong>

```python
t = (10, 20, 30)
print(t[-1])
```

<strong style="font-size : 25px; color:green">slicing</strong>

```python
# Extracting elements from tuple using slicing
t = (10, 20, 30, 40)
print(t[1:3])
```

<strong style="font-size : 25px; color:green">TypeError</strong>

```python
# Tuples are immutable and values cannot be changed
t = (10, 20, 30)
#t[0] = 100  # Gives TypeError
```

<strong style="font-size : 25px; color:green">Adding elements by creating a new tuple</strong>

```python
t = (10, 20)
t = t + (30,)
print(t)
```

<strong style="font-size : 25px; color:green">Tuple packing</strong>

```python
# Tuple packing stores multiple values into one tuple
student = "Karthik", 25, "Python"
print(student)
```

<strong style="font-size : 25px; color:green">Tuple unpacking</strong>

```python
# Tuple unpacking assigns tuple values to variables
student = ("Karthik", 25, "Python")
name, age, skill = student
print(name, age, skill)

# Swapping variables using tuple unpacking
a = 10
b = 20
a, b = b, a
print(a, b)
```

<strong style="font-size : 25px; color:green">Looping through tuple</strong>

```python
# Looping through tuple elements using for loop
numbers = (10, 20, 30)
for num in numbers:
    print(num)
```

<strong style="font-size : 25px; color:green">length of tuple using len()</strong>

```python
# Finding number of elements in tuple using len()
t = (10, 20, 30)
print(len(t))
```

<strong style="font-size : 25px; color:green">in operator</strong>

```python
# Checking whether an element exists in tuple using in operator
t = (10, 20, 30)
print(20 in t)
```

```python
# Counting occurrences of an element using count()
t = (10, 20, 10, 30)
print(t.count(10))
```

```python
# Finding first occurrence index using index()
t = (10, 20, 30)
print(t.index(20))
```

```python
# Creating nested tuples (tuple inside tuple)
t = ((1, 2), (3, 4))
print(t)
```

```python
# Joining two tuples using concatenation operator
a = (1, 2)
b = (3, 4)
print(a + b)
```

```python
# Repeating tuple elements using multiplication operator
t = (1, 2)
print(t * 3)
```

```python
# Converting list into tuple using tuple()
numbers = [10, 20, 30]
print(tuple(numbers))
```

```python
# Converting tuple into list using list()
t = (10, 20, 30)
print(list(t))
```

```python
# Tuple can be used as dictionary key because it is immutable
data = {(10, 20): "Point"}
print(data[(10, 20)])
```

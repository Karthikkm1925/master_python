<strong style="font-size: 30px; color: green;">Functions in Python</strong>

```python
 #What is a Function? : A function is a reusable block of code.

print("Without a function")
print("Welcome to Python")
print("Welcome to Python")
print("Welcome to Python")

#write once use N number of times
print("\nUsing Function")
def welcome(): #creating function
    print("Welcome to Python")
   
welcome() #calling function
'''
def → creates a function
function_name → name of the function
() → input|parameter|Formal Arguments to function, these are optional
: → starts the function body
indentation → code inside function
'''
welcome() #calling function again
welcome()
```

<strong style="font-size: 25px; color: green;">Function Without Input | Without Parameters</strong>

```python
#function to add two numbers
#values are fixed.
def addition():
    a = 10
    b = 20
    print("Sum =",a+b)
addition()
```

<strong style="font-size: 25px; color: green;">Function With Input | With Parameters</strong>

```python
#a,b are called formal parameters
def addition(a,b): #we can have one more required number of parameters
    print("Sum =",a+b)
addition(10,20) #10,20 are actual parameter | arguments | values 
```

<strong style="font-size: 25px; color: green;">Function with retun statement</strong>

```python
# 'return' sends a value back from a function.
#print = display value
#return = give value back
print("Function without Return Statement")
def add(a,b):
    print("Sum =",a+b)
add(20,30)

print("\nFunction with return statement")
def sub(a,b):
    return a-b #for any function 'return' is last statement
#when function with return statement being called returns some value 
#sub() function returns difference of two values get stored in result variable
result = sub(30,20) 
print("Difference is",result)
```

<strong style="font-size: 25px; color: green;">Function With default Parameters</strong>

```python
#Default parameter : provide default input value assigned to parameter 
def greet(name="Alan"):
    print("Hello", name)
print("This is default parameter value")
greet()#function without arguments
print("Function call with value")
greet("Lisa") #function with arguments
```

<strong style="font-size: 25px; color: green;">Function with Keyword arguments</strong>

```python
def greet(name="Alan", country = "India"):
    print("Hello I am", name,"I am from",country)
#default value
greet()
#Normal:
greet("Antony")
#Keyword:
greet(country = "USA")
greet(name="Shreehan")
```

<strong style="font-size: 25px; color: green;">Function Unlimited Arguments (*args)</strong>

```python
#Used when the number of inputs is unknown.
def total(*numbers):
    result = 0
    for n in numbers:
        result += n
    return result
print("Sum =",total(10,20,30))
```

<strong style="font-size: 25px; color: green;">Function  Keyword Unlimited Arguments (**kwargs)</strong>

```python
def details(**data):
    print(data)
details(name="Annashekh", age=52, city="Chennai")
```

<strong style="font-size: 25px; color: green;">Calling One Function Inside Another</strong>

```python
#function to calculate square of number
def square(n):
    return n*n
#function to calculate cube of number
def cube(n):
    #calling square() inside cube()
    return square(n)*n
print("Cube value =",cube(3))
```

<strong style="font-size: 25px; color: green;">Local and Global Variables</strong>

```python
#Local variable:
def test1():
    y = 10
    print(y,"is function scoped value")
test1()
#Global variable:
x = 100
def test2():
    print(x,"is global scoped")
test2()
```

<strong style="font-size: 25px; color: green;">Lambda Function</strong>

```python
#Lamda function Short one-line function.
#Normal:
def add(a,b):
    return a+b
print("Normal function result",add(10,20))
#Lambda:
add = lambda a,b: a+b
print("Lambda function result",add(10,40))
```

<strong style="font-size: 25px; color: green;">Swap Twao Numbers</strong>

```python
a = 10
b = 20

temp = a
a = b
b = temp

print("a =", a)
print("b =", b)
```

<strong style="font-size: 25px; color: green;">Using Addition and Subtraction</strong>

```python
a = 10
b = 20

a = a + b
b = a - b
a = a - b

print("a =", a)
print("b =", b)
```

<strong style="font-size: 25px; color: green;">Using Bitwise XOR</strong>

```python
a = 10
b = 20

a = a ^ b
b = a ^ b
a = a ^ b

print("a =", a)
print("b =", b)
```

<strong style="font-size: 30px; color: green;">Binary to Decimal</strong>

<strong style="font-size: 25px; color: green;">Using for loop</strong>

```python
binary = input("Enter a binary number: ")
print("You Entered",binary)
decimal = 0
power = 0

for digit in binary[::-1]:
    decimal = decimal + int(digit) * (2 ** power)
    power += 1

print("Decimal =", decimal)
```

<strong style="font-size: 25px; color: green;">Using while loop</strong>

```python
binary = int(input("Enter a binary number: "))
print("You Entered",binary)
decimal = 0
power = 0

while binary > 0:
    digit = binary % 10
    decimal = decimal + digit * (2 ** power)
    power += 1
    binary = binary // 10

print("Decimal =", decimal)
```

<strong style="font-size: 30px; color: green;">Factorial of a Number</strong>

<strong style="font-size: 25px; color: green;">Using for loop</strong>

```python
num = int(input("Enter a Number"))
print("You Entered a Number",num)
fact = 1
if num < 0:
    print("Enter Natural Numbers...")
if num ==0 or num ==1:
    print("Factorial of ",num,"is",fact)
else :
    for i in range(1, num+1):
        fact = fact * i
    print("Factorial of Number",num,"is",fact)
```

<strong style="font-size: 25px; color: green;">Using While loop</strong>

```python
num = int(input("Enter a number: "))
print("You Entered Number",num)
fact = 1
i = 1
if num < 0:
    print("Enter Natural Numbers..")
if num == 0 or num ==1:
    print("Factorial of Number",num,"is",fact)
else:
    while i <= num:
        fact = fact * i
        i += 1
    print("Factorial of", num, "is", fact)
```

<strong style="font-size: 25px; color: green;">Using Recursion</strong>

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

num = int(input("Enter a number: "))

if num < 0:
    print("Enter Natural Number..")
else:
    print("Factorial of", num, "is", factorial(num))
```

<strong style="font-size: 30px; color: green;">Extract Prime digits from given number</strong>

```python
num = int(input("Enter a number: "))
print("You Entered",num)
print("Prime Number's are ")
while num > 0:
    digit = num % 10

    if digit > 1:
        count = 0

        for i in range(1, digit + 1):
            if digit % i == 0:
                count += 1

        if count == 2:
            print(digit)

    num = num // 10
```

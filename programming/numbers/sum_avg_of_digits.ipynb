<strong style="font-size: 30px; color: green;">Sum and Average of digits in given number</strong>

<strong style="font-size: 25px; color: green;">using while loop</strong>

```python
while True:
    num = int(input("Enter a non-zero number: "))
    if num != 0:
        break
    print("0 is not allowed. Please enter a non-zero number.")
print("You Entered",num)
temp = abs(num)
sum_digits = 0
count = 0

while temp > 0:
    digit = temp % 10
    sum_digits += digit
    count += 1
    temp = temp // 10

average = sum_digits / count

print("Sum of digits =", sum_digits)
print("Average of digits =", average)
```

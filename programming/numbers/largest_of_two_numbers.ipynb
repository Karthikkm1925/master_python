<strong style="font-size: 30px; color: green;">Largest of two numbers</strong>

<strong style="font-size: 25px; color: green;">Using if-else</strong>

```python
num1 = int(input("Enter First Number"))
num2 = int(input("Enetr Second Number"))
print("You Entered",num1,"and",num2)
if num1 == num2:
    print(num1 ,"and",num2,"are same")
elif num1 > num2:
    print(num1,"is Largest")
else:
    print(num2,"is largest")
```

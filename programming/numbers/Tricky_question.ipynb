<strong style="font-size: 25px; color: green;">Add two numbers without (+) </strong>

```python
num1 = int(input("Enter first number :"))
num2 = int(input("Enter second number :"))
print("You Entered",num1," and ",num2)
print("Addition :",num1 - (-num2))
```

<strong style="font-size: 25px; color: green;">Multiply without (*)</strong>

```python
num1 = int(input("Enter first number :"))
num2 = int(input("Enter second number :"))
print("You Entered",num1," and ",num2)
result = 0
for i in range(num2):
    result += num1
print("Product :",result)
```

<strong style="font-size: 25px; color: green;">Reminder without (%)</strong>

```python
num1 = int(input("Enter first number :"))
num2 = int(input("Enter second number :"))
print("You Entered",num1," and ",num2)
while num1 >= num2:
    num1 -= num2
print("Reminder :",num1)
```

<strong style="font-size: 30px; color: green;">Check given number is Positive | Negative | Zero</strong>

<strong style="font-size: 25px; color: green;">Simple if-elif-else</strong>

```python
num = int(input("Enter a Number"))
print("You Entered",num)
if num==0:
    print("Zero")
elif num>0:
    print("Positive Number")
else:
    print("Nagative Number")
```

<strong style="font-size: 25px; color: green;">Nested if</strong>


```python
num = int(input("Enter a Number"))
print("You Entered",num)
if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive Number")
else:
    print("Negative Number")
```

<strong style="font-size: 25px; color: green;">Using Ternary (Conditional Expression)</strong>

```python
num = int(input("Enter a Number"))
print("You Entered",num)
result = "Positive Number" if num > 0 else "Negative Number" if num < 0 else "Zero"
print(result) 
```

<strong style="font-size: 25px; color: green;">Using Dictionary with Boolean Expressions</strong>

```python
num = int(input("Enter a Number"))
print("You Entered",num)
result = {num==0 : "Zero", num > 0 : "Positive Number", num < 0 : "Negative Number"}
print(result[True])
```

<strong style="font-size: 25px; color: green;">Using Match-Case</strong>

```python
num = int(input("Enter a Number"))
print("You Entered",num)
match True:
    case _ if num == 0:
        print("Zero")
    case _ if num > 0:
        print("Positive Number")
    case _ if num < 0:
        print("Negative Number")
```

<strong style="font-size: 25px; color: green;">Using a Function</strong>

```python
def check(num):
    if num == 0:
        return "Zero"
    elif num > 0:
        return "Positive Number"
    else :
        return "Negative"
num = int(input("Enter a Number"))
print("You Entered",num)
print(check(num))
```

<strong style="font-size: 25px; color: green;">Using List Indexing</strong>

```python
list1 = ["Positive", "Negative", "Zero"]
num = int(input("Enter a Number"))
print("You Entered",num)
if num == 0:
    print(list1[2])
elif num > 0:
    print(list1[0])
else :
    print(list1[1])
```

<strong style="font-size: 25px; color: green;">Using Lambda Function</strong>

```python
num = int(input("Enter a number: "))
print("You Entered",num)
check = lambda num: "Positive" if num > 0 else "Negative" if num < 0 else "Zero"
print(check(num))
```

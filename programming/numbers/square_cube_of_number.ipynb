<strong style="font-size: 30px; color: green;">Find the square root and cube root of a given number</strong>

```python
num = float(input("Enter a number: "))
print("You Entered",num)
square_root = num ** 0.5
cube_root = num ** (1/3)

print("Square Root =", square_root)
print("Cube Root =", cube_root)
```

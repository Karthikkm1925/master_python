<strong style="font-size: 30px; color: green;">Check Number is Even or Odd</strong>

<strong style="font-size: 25px; color: green;">Using Modulus Operator (%)</strong>

```python
num = int(input("Enter a Positive Integer Number"))
print("You Entered Number",num)
if(num % 2 == 0):
    print(num,"is Even Number")
else:
    print(num,"is Odd Number")
```

<strong style="font-size: 25px; color: green;">Using Bitwise AND Operator (&)</strong>

```python
num = int(input("Enter a Positive Integer Number"))
print("You Entered Number",num)
if num & 1 == 0:
    print(num,"is Even Number")
else:
    print(num,"is Odd Number")
```

<strong style="font-size: 25px; color: green;">Using Integer Division (//)</strong>

```python
num = int(input("Enter a Positive Integer Number"))
print("You Entered",num)
if (num // 2) * 2 == num:
    print(num,"is Even Number")
else:
    print(num,"is Odd Number")
```

<strong style="font-size: 25px; color: green;">Using Bitwise Shift Operator (>>,<<)</strong>

```python
num = int(input("Enter a Positive Number"))
print("You Entered",num)
if (num >> 1)<<1 == num:
    print(num,"is Even Number")
else:
    print(num,"is Odd Number")
```

<strong style="font-size: 25px; color: green;">Using XOR Operator (^)</strong>

```python
num = int(input("Enter a Positive Number"))
print("You Entered",num)
if (num^1) == num +1:
    print(num,"is Even Number")
else:
    print(num,"is Odd Number")
```

<strong style="font-size: 25px; color: green;">Using Recursion</strong>

```python
def check(num):
    if num == 0:
        return "Even"
    if num == 1:
        return "Odd"
    return check(num - 2)
num = int(input("Enter a Positive Integer"))
print("You Entered",num)
print(num,"is",check(num),"Number")
```

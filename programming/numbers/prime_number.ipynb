<strong style="font-size: 30px; color: green;">Check given number is prime?</strong>

<strong style="font-size: 25px; color: green;">Using for loop</strong>

```python
num = int(input("Enter a Number"))
print("You Entered",num)
if num <=1:
    print("Enter number greater than 1")

else:
    for i in range(2, num):
        if num % i  ==0:
            print(num,"Not a Prime Number")
            break
    else:
        print(num,"is a Prime Number")
            
```

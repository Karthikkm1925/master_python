<strong style="font-size: 30px; color: green;">Factorial of each digit in given number</strong>

```python
num = int(input("Enter a number: "))
print("You Entered",num)
n = num
while n > 0:
    digit = n % 10

    fact = 1
    i = 1
    while i <= digit:
        fact = fact * i
        i += 1

    print(f"Factorial of {digit} = {fact}")

    n = n // 10
```

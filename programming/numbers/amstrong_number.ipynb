<strong style="font-size: 30px; color: green;">Check Given Number is Amstrong Number?</strong>

```python
num = int(input("Enter a number: "))

original = num
sum = 0
digits = 0

# Count number of digits
temp = num
while temp > 0:
    digits += 1
    temp = temp // 10

# Calculate Armstrong value
temp = num
while temp > 0:
    digit = temp % 10
    sum = sum + (digit ** digits)
    temp = temp // 10

# Check condition
if sum == original:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

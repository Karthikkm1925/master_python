<strong style="font-size: 30px; color: green;">Happy Number or Not?</strong>

```python
num = int(input("Enter a number: "))
print("You Entered",num)
original = num
seen = set()

while num != 1 and num not in seen:
    seen.add(num)

    sum = 0

    while num > 0:
        digit = num % 10
        sum = sum + (digit * digit)
        num = num // 10

    num = sum

if num == 1:
    print(original, "is a Happy Number")
else:
    print(original, "is not a Happy Number")
```

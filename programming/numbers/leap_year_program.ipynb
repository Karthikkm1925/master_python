<strong style="font-size: 30px; color: green;">Leap Year Program</strong>

<strong style="font-size: 25px; color: green;">Using Nested if</strong>

```python
year = int(input("Enter a year: "))
if year <1000 :
    print("Enter valid year")
if year % 400 == 0:
    print(year, "is a Leap Year")
elif year % 100 == 0:
    print(year, "is Not a Leap Year")
elif year % 4 == 0:
    print(year, "is a Leap Year")
else:
    print(year, "is Not a Leap Year")
```

<strong style="font-size: 25px; color: green;">Using Combined Condition</strong>

```python
year = int(input("Enter a year: "))
if year <1000 :
    print("Enter valid year")
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year")
else:
    print(year, "is Not a Leap Year")
```

<strong style="font-size: 25px; color: green;">Using Boolean Variable</strong>

```python
year = int(input("Enter a year: "))
if year <1000 :
    print("Enter valid year")
is_leap = (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)

if is_leap:
    print(year, "is a Leap Year")
else:
    print(year, "is Not a Leap Year")
```

<strong style="font-size: 25px; color: green;">Using Ternary Operator</strong>

```python
year = int(input("Enter a year: "))
if year <1000 :
    print("Enter valid year")
result = "Leap Year" if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0) else "Not a Leap Year"

print(year, "is", result)
```

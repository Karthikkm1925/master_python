<strong style="font-size: 30px; color: green;">Count number of digits in given Number</strong>

<strong style="font-size: 25px; color: green;">Using while loop</strong>

```python
num = input("Enter a Number")
print("You Entered",num)
if num == 0:
    count = 1
else:
    if num < 0:
        num = -(num)
    count = 0
    while num > 0:
        count+=1
        num//=10
print("Number has Total",count,"digit")
#note: if prefix has digit 0, not taken into count
#e.g., 0123 = 123, count = 123
```

<strong style="font-size: 30px; color: green;">Digital Root or Repeated sum of digits</strong>

<strong style="font-size: 25px; color: green;">Using while loop</strong>

```python
num = int(input("Enter a number: "))
print("You Entered",num)
while num >= 10:
    total = 0
    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10
    num = total

print("Single digit:", num)
```
